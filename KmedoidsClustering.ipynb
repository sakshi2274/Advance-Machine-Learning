```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset
data = np.array([
    [2,6],
    [3,4],
    [3,8],
    [4,7],
    [6,2],
    [6,4],
    [7,3],
    [7,4],
    [8,5],
    [7,6]
])
```

```python
k = 2
medoids = data[[0, 4]]
```

```python
def distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2) ** 2))
```

```python
clusters = [[] for _ in range(k)]

for point in data:
    distances = [distance(point, medoid) for medoid in medoids]
    cluster_index = np.argmin(distances)
    clusters[cluster_index].append(point)
```

```python
clusters = [np.array(cluster) for cluster in clusters]
```

```python
colors = ['blue','green']

for i in range(k):
    plt.scatter(clusters[i][:,0], clusters[i][:,1], color=colors[i])

plt.scatter(medoids[:,0], medoids[:,1], color='red', marker='X', s=200)

plt.title("K-Medoids Clustering")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
